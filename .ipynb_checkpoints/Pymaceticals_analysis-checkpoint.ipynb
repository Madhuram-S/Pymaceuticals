## Pymaceuticals - Analysis on drug-based, anti-cancer study 
In this study 250  mice were treated through a variety of drug regimes over the course of 45 days. Their physiological responses were then monitored over the course of that time. The objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare. 

### Analysis process
The below 4 metrics were analysed to determine the efficacy of the drugs:-
- Rate of change of tumor volume over 45 days
- % change of tumor volume in 45 days
- Rate of increase / decrease of metastatic sites over the treatment
- Survival rate of mice over the treatment of 45 days

In order to account for variability within the sample, Standard Error of Mean(SEM) was also determined and plotted

Scatter plots are created to produce graphical output and to see the trends of efficacy

### Observations
- Most Effiencient : Campomulin outperform the other three drugs in comparison and is the most efficient in terms tumor reduction (reduced by 19%) and better survival rates (84%). It does not stop the spread the tumor but controls to an extent
- Least Efficient : Ketapril is the least efficient with poor survival rate, high % of change in tumor volume and increased growth of metastatic sites
- The other two drugs, Infubinol and Placebo, are not effective either with % of growth in tumor volume 46% and 51% respectively. Infubinol has worst survival rate(36%) and Placebo has a higher rate of metatstaic spread

```python
# invoke magic function to produce interactive plotting
%matplotlib inline
```

```python
# Import all required dependencies and packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# pre-set the global configurations and constants for this analysis
dataFile_path = 'data'
outputImg_path = "Images"

# Drugs that are analysed
drugs_cmp = ['Capomulin','Infubinol','Ketapril','Placebo']

# set plot constants, color scheme, markers and other aesthetic parameters
m_size = 6
m_color = {'Capomulin' : 'r', 'Infubinol' : 'b', 'Ketapril' : 'g', 'Placebo' : 'black'}
m_type = {'Capomulin' : 'o', 'Infubinol' : '^', 'Ketapril' : 's', 'Placebo' : 'd'}
l_type = (0, (1, 1))
l_width = 1.5
```

```python
# load files 
# File 1 - mouse_drug_data.csv : has the drug administered information mice wise
# File 2 - clinicaltrial_data.csv : has the data collected over the trial for each mice

files_to_load = [os.path.join(dataFile_path,'mouse_drug_data.csv'),os.path.join(dataFile_path,'clinicaltrial_data.csv')]

drug_DF, trial_DF = [pd.read_csv(f) for f in files_to_load]

merged_DF = pd.merge(drug_DF, trial_DF, how = 'outer', on = ['Mouse ID'])
```

```python
# check if data is clean else drop na rows
clinicalDF = merged_DF if (merged_DF.isnull().values.sum() != 0) else merged_DF.dropna(how = 'any')
clinicalDF.head()
```

```python
# Create grouped dataframe by drug and timpoint that will be used throughout the analysis
# raw data is grouped by Drug and Timepoint
clinical_GDF = clinicalDF.groupby(['Drug', 'Timepoint'])
```

```python
# Derive the below summary values for Tumor growth and Metastatic sites
# values calculated are mean & standard error

# Mean and SEM is calculated for tumor grwoth and metastatic site increase
clinical_SDF = pd.DataFrame(clinical_GDF.agg(['mean', 'sem']))

# column names are changed to meaningful values
clinical_SDF.columns = clinical_SDF.columns.map('_'.join)
clinical_SDF.reset_index(inplace = True)
clinical_SDF.head()
```

###  Tumor Response to Treatment
We find mean tumor volume and tumor volume SEM

```python
# Disply the list of mean Tumor volume (mm3) and SEM for tumor vol.
clinical_SDF.loc[:, ['Drug','Timepoint','Tumor Volume (mm3)_mean','Tumor Volume (mm3)_sem']].head(20)
```

```python
# Display how the tumor volume changes over time for each drug and timepoint as excel type display
clinical_SDF.pivot('Timepoint','Drug', 'Tumor Volume (mm3)_mean')
```

```python
# Generate plot for how the tumor growth has happened for the four drugs (Capomulin, Infubinol, Ketapril, and Placebo)
fig, ax = plt.subplots(figsize = (7,5))

ax_hndls = []

for drug in drugs_cmp:
    ax_hndls.append(ax.errorbar('Timepoint','Tumor Volume (mm3)_mean', 'Tumor Volume (mm3)_sem',
                                data = clinical_SDF.loc[ clinical_SDF['Drug'] == drug, :],
                                marker = m_type[drug],
                                color = m_color[drug],
                                linewidth = l_width,
                                linestyle = l_type,
                                markersize = m_size,
                                label = drug                                                               
                               )
    )
    
# Apply grid, display legend, title and axis labels
plt.grid()
plt.legend(loc = 'best')
plt.title('Tumor Response to treatment')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')

#save plot to png Image file
plt.savefig(os.path.join(outputImg_path,'TumorResponse.png'))

#display plot as inline
plt.show()
    
```

### Metastatic Response to Treatment

```python
# Disply the list of mean Metastatic sites and SEM for metastatic sites
clinical_SDF.loc[:, ['Drug','Timepoint','Metastatic Sites_mean','Metastatic Sites_sem']].head(20)
```

```python
# Display the mean of metastatic list so all drugs information can be viewed as single frame
clinical_SDF.pivot('Timepoint','Drug','Metastatic Sites_mean')
```

```python
# Generate plot for spread of metastatic sites when using the four drugs (Capomulin, Infubinol, Ketapril, and Placebo)
fig, ax = plt.subplots(figsize = (7,5))

ax_hndls = []

for drug in drugs_cmp:
    ax_hndls.append(ax.errorbar('Timepoint','Metastatic Sites_mean','Metastatic Sites_sem',
                                data = clinical_SDF.loc[ clinical_SDF['Drug'] == drug, :],
                                marker = m_type[drug],
                                color = m_color[drug],
                                linewidth = l_width,
                                linestyle = l_type,
                                markersize = m_size,
                                label = drug                                                               
                               )
    )

# Apply grid, display legend, title and axis labels
plt.grid()
plt.legend(loc = 'best')
plt.title('Metastatic Spread during treatment')
plt.xlabel('Time (days)')
plt.ylabel('Met. Sites')

#save the plot as a png image
plt.savefig(os.path.join(outputImg_path,'MetastaticSpread.png'))
#show plot on console
plt.show()
    
```

### Survival Rates
Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)

```python
# using the groupd DF created above, create a DF with two columns - 
#Surviving Mouse count and Survival Rate inaddition to Drug and Timepoint

survival_SDF = pd.DataFrame(clinical_GDF['Mouse ID'].count())
survival_SDF['Survival rate'] = \
                        (survival_SDF['Mouse ID'] / (survival_SDF.groupby('Drug', sort = False)['Mouse ID'].max())) *100

survival_SDF.rename(columns = {'Mouse ID' : 'Surviving Count'}, inplace = True)
survival_SDF.reset_index(inplace = True)
survival_SDF.head()
```

```python
# Display the Rate of survival of mice over the 45 days for the four drugs
fig, ax = plt.subplots(figsize = (7,5))

ax_hndls = []

for drug in drugs_cmp:
    ax_hndls.append(ax.plot('Timepoint','Survival rate',
                                data = survival_SDF.loc[survival_SDF['Drug'] == drug, :],
                                marker = m_type[drug],
                                color = m_color[drug],
                                linewidth = l_width,
                                linestyle = l_type,
                                markersize = m_size,
                                label = drug                                                               
                               )
    )

# Apply grid, display legend, title and axis labels
plt.grid()
plt.legend(loc = 'best')
plt.title('Survival during treatment')
plt.xlabel('Survival rate (%)')
plt.ylabel('Time (days)')

#save the plot as a png image
plt.savefig(os.path.join(outputImg_path,'SurvivalRate.png'))
#show plot on console
plt.show()
```

### Tumor Change over 45 days for the four drugs

```python
# We will use the already created clinical_SDF dataframe and extract only the Drug, timepoint and tumor volume mean
tumor_chng_SDF = clinical_SDF.loc[clinical_SDF['Timepoint'].isin([0,45]),["Drug", "Timepoint", 'Tumor Volume (mm3)_mean'] ]

#Calculate % change
tumor_chng_SDF = tumor_chng_SDF.pivot('Drug','Timepoint','Tumor Volume (mm3)_mean')
tumor_chng_SDF['Percent Change'] = ((tumor_chng_SDF[45] - tumor_chng_SDF[0])/ tumor_chng_SDF[0])*100
tumor_chng_SDF.reset_index(inplace = True)

tumor_chng_SDF
```

```python
# filter the above DF for only the required drugs for graphing purpose
tumor_chng_filt = tumor_chng_SDF.loc[tumor_chng_SDF['Drug'].isin(drugs_cmp),:]

# plot the change in tumor volume only for the four drugs (Capomulin, Infubinol, Ketapril, Placebo)
fig, ax = plt.subplots(figsize = (7,5))

lgnd_hndl = ax.bar('Drug','Percent Change',
                    data = tumor_chng_filt,
                    width = -1,
                    align = 'edge',
                    tick_label = drugs_cmp,
                    color = ['g' if(c) else 'r' for c in list(tumor_chng_filt['Percent Change'] <= 0)]
                    
                   )
ax.grid()
# set y limit
ax.set_ylim(-30,tumor_chng_filt['Percent Change'].max()+10, 20 )

#set data labels for each bar
for i, v in enumerate(tumor_chng_filt['Percent Change']):
    ax.text(x = i - 0.75, y = 5 if (v >=0) else -8, 
            s = (f'{round(v,0)}%'), color='w', fontweight='bold')

# set title and labels

ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change over 45 day treatment')

plt.savefig(os.path.join(outputImg_path, 'Percent_TumorChng.png'))
    
plt.show()
```
